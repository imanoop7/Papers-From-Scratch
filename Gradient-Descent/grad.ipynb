```python
import numpy as np
```

```python
# equation of line -> yhat = mx +c or yhat = mx + b
#MSE = ((yhat-y)**2)/N
X= np.random.rand(10,1)
y = 2*X + np.random.rand()
```

```python
w= 0.0 #slope or weights
b=0.0 #intercept or bias
learning_rate = 0.01
```

```python
def grad_descend(X,y,w,b,learning_rate):
    dw = 0.0
    db= 0.0
    N = X.shape[0]

    for xi, yi in zip(X,y):
        dw = -2*xi*(yi-(w*xi+b))
        db = -2*(yi-(w*xi+b))

    w = w - learning_rate *(1/N) *dw
    b = b- learning_rate *(1/N) *db

    return w,b
```

```python
# Prediction function
def predict(X, w, b):
    return X * w + b
```

```python
for epoch in range(1000):
    w,b=grad_descend(X,y,w,b,learning_rate)
    yhat = predict(w,X,b)
    loss = np.divide(np.sum((y-yhat)**2,axis=1),X.shape[0])
    print(f"{epoch} loss is {loss}, weight w:{w}, bias{b}")
    
```
